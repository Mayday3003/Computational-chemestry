```python
#Importación de método para poder leer el archivo, en este  caso .cif
from ase.io import read

from ase.visualize import view
```

```python

# Se lee el archivo .cif usando ase.io.read
Ag2S2O7 = read('Ag2S2O7.cif')

'''
Como ya está cargado y el se ha creado la instancia de la mólecula automaticamente al usar
el método read, podemos pasar a la visualización
'''

#Usando el método view con la molécula creada luego de subir el archivo definirla como Ag2S2O7
view(Ag2S2O7, viewer='x3d')
```
